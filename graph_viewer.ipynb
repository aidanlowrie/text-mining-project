```python
def get_df_with_relation(relation):
    columns = ['source', 'relation','target']
    new_df = pd.DataFrame(columns=columns) 
    for index, row in df.iterrows():
        if row['relation']==relation:
            new_df.loc[len(new_df)] = [row[mp[i]] for i in range(3)]
            
    return new_df        
```

```python
def get_df_with_word(word):
    columns = ['source', 'relation','target']
    new_df = pd.DataFrame(columns=columns) 
    for index, row in df.iterrows():
        if row['source']==word or row['target']==word:
            new_df.loc[len(new_df)] = [row[mp[i]] for i in range(3)]
    return new_df        
```

```python
G = nx.from_pandas_edgelist(get_df_with_relation('follows'))
pos = nx.spring_layout(G)
nx.draw(G,node_size= 25)
```

```python
G = nx.from_pandas_edgelist(get_df_with_word('protein') )
pos = nx.spring_layout(G)
nx.draw(G,node_size= 25, with_labels=True)
```
